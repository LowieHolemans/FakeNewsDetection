```python
def get_jaccard_sim(str1, str2):
   a = set(str1.split())
   b = set(str2.split())
   c = a.intersection(b)
   return float(len(c)) / (len(a) + len(b) - len(c))
```

```python
def calc_avg_sim(data): 
    total = 0 
    teller = 0
    for i in range(len(data['text'])): 
        if i%100 == 0: 
            print(i)
        for j in range(1,len(data['text'])):
            total = total + get_jaccard_sim(data.iloc[i,1],data.iloc[j,1])
            teller = teller + 1 

    avg_sim = total/teller

    return avg_sim
```

```python
total = (calc_avg_sim(df_0)+calc_avg_sim(df_1)+calc_avg_sim(df_2)+calc_avg_sim(df_3)+calc_avg_sim(df_4)+calc_avg_sim(df_5)+calc_avg_sim(df_6)+calc_avg_sim(df_7)+calc_avg_sim(df_8)+calc_avg_sim(df_9))/10
print(total)
```
